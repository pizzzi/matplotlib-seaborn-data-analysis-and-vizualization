```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Coffee Production Stats

* Bar chart of the top 10 coffee producers by total production 


```python
coffee_production=pd.read_csv("../Data/coffee/total-production.csv").T
```

```python
coffee_production.columns= coffee_production.iloc[0]
coffee_production=coffee_production.drop("total_production")
coffee_production.head()
```

```python
top10_coffee_producers=coffee_production.sum(axis=0).sort_values(ascending=False).iloc[:10]
top10_coffee_producers
```

```python
fig,ax =plt.subplots(figsize=(8,8))
ax.bar(top10_coffee_producers.index, top10_coffee_producers)
labels=["BRA", "VIE","COL","INDO","ETH", "IND", "MEX", "GUA", "HON","UGA"]
ax.set_xticklabels(labels);
ax.set_title("Top Coffee Producing Nations 1990-2018")
ax.set_ylabel("Production (Millions 60kg Bags)")
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
```

# Top 5 over time

* Line chart with the top 5 nations over time.

```python
top5 = coffee_production.loc[:,["Brazil","Viet Nam","Colombia","Indonesia","Ethiopia"]]
top5.index = top5.index.astype("datetime64[ns]")
top5.head()
```

```python
fig,ax =plt.subplots(figsize=(8,8))
ax.plot(top5.index, coffee_production["Brazil"] / 1000, label="Brazil")
ax.plot(top5.index, coffee_production["Viet Nam"] / 1000, label="Viet Nam")
ax.plot(top5.index, coffee_production["Colombia"] / 1000, label="Colombia")
ax.plot(top5.index, coffee_production["Indonesia"] / 1000, label="Indonesia")
ax.plot(top5.index, coffee_production["Ethiopia"] / 1000, label="Ethiopia")
ax.set_title("Top 5 Coffee Producing Nations 1990-2018")
ax.set_ylabel("Production (Millions 60kg Bags)")
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.legend()
```

# COMPOSITION OVER TIME

```python
coffee_production=pd.read_csv("../Data/coffee/total-production.csv").T
coffee_production.columns= coffee_production.iloc[0]
coffee_production=coffee_production.drop("total_production")
coffee_production.head()
```

```python
coffee_production["Rest Of World"]=coffee_production.drop(["Brazil"], axis=1).sum(axis=1)
```

```python
pie_data=coffee_production.loc[:,["Brazil","Rest Of World"]]
pie_data
```

```python
fig, ax=plt.subplots()

ax.pie(pie_data.loc["1990"][::-1], startangle=90, colors=["white","green"])

hole= plt.Circle((0,0), 0.70,fc="white")
fig = plt.gcf()
fig.gca().add_artist(hole)

plt.text(0,0, f"{round((pie_data.loc['1990', 'Brazil'] / pie_data.loc['1990'].sum()*100))}%", ha="center", va="center", fontsize=42)
ax.set_title("Brazil Share of Global Production 1990");
```

```python
fig, ax=plt.subplots()

ax.pie(pie_data.loc["2018"][::-1], startangle=90, colors=["white","green"])

hole= plt.Circle((0,0), 0.70,fc="white")
fig = plt.gcf()
fig.gca().add_artist(hole)

plt.text(0,0, f"{round((pie_data.loc['2018', 'Brazil'] / pie_data.loc['2018'].sum()*100))}%", fontsize=42, ha="center", va="center" )
ax.set_title("Brazil Share of Global Production 2018");
```

# Coffee Consumption Vs. Prices

Dual axis bar chart of price vs consumption totals for these countries.

```python
consumption = pd.read_csv("../Data/coffee/imports.csv").set_index("imports").mean(axis=1)


consumption
```

```python
prices = pd.read_csv("../Data/coffee/retail-prices.csv").set_index("retail_prices").mean(axis=1)

prices.head()
```

```python
pri_con=pd.DataFrame(prices).merge(pd.DataFrame(consumption), left_index=True, right_index=True, how="inner").sort_values("0_y", ascending=False).rename(columns={
    "0_x": "Price",
    "0_y": "Imports"
})
pri_con.head()
```

```python
fig, ax= plt.subplots(figsize=(14,8))
width= 0.35
fig.suptitle("Prices Paid Vs. Consumption of Select Importing Nations")
x=np.arange(len(pri_con))
ax.bar((x-width/2), pri_con["Imports"], color="blue", width=width, label="Consumption")
ax.set_ylabel("Consumption (k 60kg bags)")
ax2=ax.twinx()
ax2.bar((x+width/2), pri_con["Price"], color="orange", width=width, label="Price")
ax2.set_ylabel("Average Price Paid Per Bag (USD)")

plt.xticks(x, pri_con.index, fontsize=7)

fig.legend()
```
